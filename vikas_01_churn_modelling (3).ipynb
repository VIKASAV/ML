## Cross Validation

```python
#perform cross validation on the xtrain & y train
from sklearn.model_selection import StratifiedKFold
#perform cross validation
skf= StratifiedKFold(n_splits=3, random_state=None, shuffle=False)
```

```python
from sklearn.naive_bayes import GaussianNB
print('NB Classifier')
for c in [1,2,3]:

    cv_score_mean=0
    for train_index, test_index in skf.split(trainx_scale, trainy):
        print('Train:',train_index,'Test:',test_index)
        trainxcv,val_x_cv=trainx_scale[train_index], trainx_scale[test_index]
        trainycv,val_y_cv=trainy.iloc[train_index], trainy.iloc[test_index]
        classifier= GaussianNB() # creating model
        classifier.fit(trainxcv,trainycv) #training the model
        y_pred = classifier.predict(val_x_cv) #Training a Model
        cv_score=roc_auc_score(y_true=val_y_cv, y_score=y_pred[:])
        cv_score_mean=cv_score_mean+cv_score
        print('ROC-AUC score', cv_score_mean/3)
```

# Why cross validation performed well ?

## Cross Validation is usually a very good way to measure an accurate performance. This involves simply repeating the cross-validation procedure multiple times and reporting the mean result across all folds from all runs.
